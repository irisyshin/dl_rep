irisdataset load

```python
#iris dataset load
from sklearn import datasets
iris = datasets.load_iris()
X = iris.data
y = iris.target
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

신경망 구조 생성

```python
from tensorflow.keras.layers import Input, Dense, Dropout, PReLU, Activation, ELU
from tensorflow.keras.models import Sequential

model = Sequential()
model.add(Input(shape=(4,))) # 입력
model.add(Dense(50, activation='sigmoid'))
model.add(Dense(50, activation=PReLU()))
model.add(Dropout(0.2))
model.add(Dense(30))
model.add(Activation(ELU(alpha=1)))
model.add(Dense(3, activation='softmax'))
model.summary()
```

가중치 초기화

```python

from tensorflow.keras.layers import Input, Dense, Dropout, PReLU, Activation, ELU
from tensorflow.keras.models import Sequential
from tensorflow.keras.initializers import RandomNormal

model1 = Sequential()
model1.add(Input(shape=(4,)))
model1.add(Dense(50, activation='sigmoid', kernel_initializer='he_normal'))
model1.add(Dense(50, activation=PReLU()))
model1.add(Dense(30, activation='sigmoid', kernel_initializer=RandomNormal(mean=0, stddev=0.1)))
model1.add(Dense(3, activation='softmax'))
model1.summary()
```

배치정규화

```python
from tensorflow.keras.layers import BatchNormalization
model2 = Sequential()
model2.add(Input(shape=(4,)))
model2.add(Dense(50, activation='sigmoid'))
model2.add(BatchNormalization())
model2.add(Dense(50, activation='sigmoid'))
model2.add(BatchNormalization())
model2.add(Dense(30, activation='sigmoid'))
model2.add(BatchNormalization())
model2.add(Dense(3, activation='softmax'))
model2.summary()
```

컴파일

```python
# model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
) # 들어가는 값은 상황에 따라 바꾸기
```

```python
# model2
model2.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
) # 들어가는 값은 상황에 따라 바꾸기
```

```python
history = model.fit(X_train, y_train, epochs=300, verbose=1)
```

```python

import pandas as pd

redwine = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv", sep=";")
X, y = redwine.iloc[:, :-1], redwine.iloc[:, -1]
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
y_train = y_train - y.min()
y_test = y_test - y.min()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense, Dropout, PReLU, Activation, ELU

# 신경망 구조 생성
model = Sequential()

# 입력층 + Hidden 1
model.add(Input(shape=(11,)))   # 입력 feature 11개
model.add(Dense(100, activation='sigmoid'))
model.add(Dropout(0.2))

# Hidden 2
model.add(Dense(200, activation='relu'))
model.add(Dropout(0.4))

# Hidden 3
model.add(Dense(50, activation='tanh'))
model.add(Dropout(0.1))

# 출력층
model.add(Dense(6, activation='softmax'))

# 모델 요약 출력
model.summary()
```

```python
# 가중치 초기화
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense, Dropout, PReLU, Activation, ELU
from tensorflow.keras.initializers import RandomNormal

# 신경망 모델 정의
model = Sequential()

# 입력층 + Hidden 1
model.add(Input(shape=(11,)))   # 입력 feature 11개
model.add(Dense(100, activation='sigmoid', kernel_initializer='he_normal'))
model.add(Dropout(0.2))

# Hidden 2
model.add(Dense(200, activation='relu', kernel_initializer='he_normal'))
model.add(Dropout(0.4))

# Hidden 3
model.add(Dense(50, activation='tanh', kernel_initializer=RandomNormal(mean=0, stddev=0.1)))
model.add(Dropout(0.1))

# 출력층
model.add(Dense(6, activation='softmax'))

# 모델 요약
model.summary()
```

```python
# 신경망 모델 정의
model2 = Sequential()

# 입력층 + Hidden 1
model2.add(Input(shape=(11,)))   # 입력 feature 11개
model2.add(Dense(100, activation='sigmoid', kernel_initializer='he_normal'))
model2.add(BatchNormalization())
model2.add(Dropout(0.2))

# Hidden 2
model2.add(Dense(200, activation='relu', kernel_initializer='he_normal'))
model2.add(BatchNormalization())
model2.add(Dropout(0.4))

# Hidden 3
model2.add(Dense(50, activation='tanh', kernel_initializer=RandomNormal(mean=0, stddev=0.1)))
model2.add(BatchNormalization())
model2.add(Dropout(0.1))

# 출력층
model2.add(Dense(6, activation='softmax'))

# 모델 구조 요약
model2.summary()
```

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
model1.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
model2.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
model.fit(X_train, y_train, epochs=300, verbose=1)
```
