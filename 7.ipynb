```python
from ultralytics import YOLO
# 다운로드 받은걸로 객체 받기
model = YOLO('yolov8n.pt')
# 모델에 바로 이미지 주기
data_path = r"datasets\Num\data.yaml"
model.train(data=data_path, epochs=50, imgsz=640)
```

```python
# 하이퍼 파라미터 찾기 실습
from ultralytics import YOLO
model = YOLO(r'runs\detect\train14\weights\best.pt')
model.tune( 
    data=r"datasets\Num\data.yaml",
    epochs=20,
    imgsz=640,
    iterations=20, # 탐색 횟수(예: 25번의 실험) 파라미터조합
    val=True
)
```

```python
import yaml

# 데이터셋 경로 설정
data_path =r"datasets\Num\data.yaml"
# 하이퍼파라미터 yaml 파일 불러오기
with open(r"runs\detect\tune5\best_hyperparameters.yaml", "r", encoding="utf-8") as f:
    hyp_dict = yaml.safe_load(f)
# 모델 훈련
model.train(data=data_path, epochs=30, imgsz=640, batch=16, **hyp_dict)
```
